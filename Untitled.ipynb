```python
from vaderSentiment.vaderSentiment import sentiment as vaderSentiment 
from Tkinter import *
import pandas as pd
from nltk import tokenize
```

```python
df = pd.read_csv("awards.csv")
```

```python
for index, row in df.iterrows():
    print index
    sentences = tokenize.sent_tokenize(row["text"])
    sumComp = 0
    sumPos = 0
    sumNeg = 0
    tot = len(sentences)
    pos = 0
    neg = 0
    neu = 0
    compPos = 0
    compNeg = 0
    compNeutral = 0
    for sentence in sentences: 
        vs = vaderSentiment(sentence)
        sumComp += vs['compound']
        sumPos += vs['pos']
        sumNeg += vs['neg']
       
    df.loc[index,'sentiment'] = sumComp / tot
```
